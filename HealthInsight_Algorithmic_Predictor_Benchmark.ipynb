```python
# Import necessary libraries for data manipulation, visualization, and machine learning
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Load the diabetes dataset from a CSV file into a pandas DataFrame
df=pd.read_csv('diabetes.csv')
```

```python
# Display the entire DataFrame to inspect its contents
df
```

```python
# Separate features (x) and target variable (y)
# 'Outcome' is the target variable indicating diabetes presence
x=df.drop('Outcome',axis=1)
y=df['Outcome']

# Import train_test_split for splitting data into training and testing sets
from sklearn.model_selection import train_test_split

# Split the dataset into training (80%) and testing (20%) sets
# random_state is set for reproducibility
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=57)
```

```python
# Initialize the K-Nearest Neighbors classifier with 3 neighbors
knn_model=KNeighborsClassifier(n_neighbors=3)

# Initialize the Gaussian Naive Bayes classifier
nb_model=GaussianNB()

# Initialize the Logistic Regression classifier with a high max_iter for convergence
lr_model=LogisticRegression(max_iter=1000)
```

```python
# Train the KNN model using the training data
knn_model.fit(x_train,y_train)

# Train the Naive Bayes model using the training data
nb_model.fit(x_train,y_train)

# Train the Logistic Regression model using the training data
lr_model.fit(x_train,y_train)
```

```python
# Make predictions on the test set using the trained KNN model
y_pred_knn=knn_model.predict(x_test)

# Make predictions on the test set using the trained Naive Bayes model
y_pred_nb=nb_model.predict(x_test)

# Make predictions on the test set using the trained Logistic Regression model
y_pred_lr=lr_model.predict(x_test)
```

```python
# Calculate and print the accuracy score for the KNN model
print("KNN:-",accuracy_score(y_test,y_pred_knn))

# Calculate and print the accuracy score for the Naive Bayes model
print("NB:-",accuracy_score(y_test,y_pred_nb))

# Calculate and print the accuracy score for the Logistic Regression model
print("Logistic Regression:-",accuracy_score(y_test,y_pred_lr))
```

```python
# List to store accuracy scores for KNN across different random states
knn_accuracy_results = []

# Loop 100 times, varying the random_state for train-test split
for i in range(100):
  # Split data with the current random_state
  x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
  # Train KNN model
  knn_model.fit(x_train,y_train)
  # Make predictions
  y_pred = knn_model.predict(x_test)
  # Calculate accuracy
  accuracy = accuracy_score(y_test, y_pred)
  # Store the random state and its corresponding accuracy
  knn_accuracy_results.append({'random_state': i, 'accuracy': accuracy})
```
